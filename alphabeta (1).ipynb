```python
#######################################
# Alpha Beta Pruning
#
# Srijith
# Lannisters don't act like fools
# It is what it is
#
# - This is a really basic implementation.
# -
#
#######################################

#######################################
# Constants

MAX = 99999
MIN = -99999

#######################################
# Game tree  : Change the list "Game_tree" to test a different game
'''
                       o MAX
                       |
           ____________|_________
          |                      |
    ______o________           ___o___     MIN
   |      |        |         |       |
   7    __o___    _o_        12    __o__       MAX
       |  |  |   |   |            |     |
       3  8  1   15  9           17     13
'''

Game_tree = [[4, [7,9,8], 8], [3, 8, [7,[11, 10, 9],[8, 10, 2]]], [[[3,6,4],2,6], [[9,2,1],5,7,[1,4,3]], 9, [6, 4, 5]]]


#######################################
# Function definitions

def minmax_decision(tree):
  '''
  input: game tree (as a python list)
  return: value of optimal outcome
  '''

  # long implementation (as in the slides)
  '''
  max = MIN

  for i in tree:
    v = minmax_value(i,"min")
    if (v > max):
      max = v

  return max
  '''

  alpha = -9999
  beta = 9999

  # compact implementation:

  return minmax_value(tree,"max", alpha, beta)


def minmax_value(tree,node_type, alpha, beta ,indent=""):
  '''
  input: game tree and the type of its root : "max" or "min"
  return: value of optimal outcome
  '''

  if (type(tree)==int):      # Leaf node
    print(indent+"CALL",node_type,"returns ",tree)
    return tree

  elif (node_type == "max"): # "tree"'s root is a MAX node


    print(indent+"CALL max "+str(tree))

    for i in tree:
      v = minmax_value(i,"min", alpha, beta,indent+"  ")
      if (v>alpha):
        alpha = v
      if(alpha >= beta):
        print("Breaking the Alpha >= Beta ", alpha, beta, "After Executing the node ",v, "It breaks at the parent")
        return alpha

    print(indent+"  - max",tree,"returns "+str(alpha))
    return alpha

  elif (node_type == "min"): # "tree"'s root is a MIN node

    print(indent+"CALL min "+str(tree))

    for i in tree:
      v = minmax_value(i,"max",alpha, beta, indent+"  ")
      if (v<beta):
        beta = v
      if(alpha >= beta):
        print("Breaking the Alpha >= Beta ", alpha, beta,  "After Executing the node ",v, "It breaks at the parent")
        return beta

    print(indent+"  - min",tree,"returns "+str(beta))
    return beta

#######################################
# Test

minmax_decision(Game_tree)
```
